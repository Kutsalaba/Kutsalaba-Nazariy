### **Завдання 1 (2 бали)**

Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
# Import modules
import math


def check(n):
    '''
    :param n: degree of polynomial;
    :return: int,degree of polynomial.
    This function checks the correctness of the input.
    '''
    try:
        n = int(input("Enter the degree of the polynomial: "))
        if int(n) >= 0:
            return n
        else:
            raise ValueError
    except ValueError:
        print("Error, incorrect data entered!")
        return check(n)
numer = check('')

def new_num(n):
    '''
    :param n: degree of polynomial;
    This function calculates binomial coefficients for binomials of all degrees.
    '''
    yield 1
    for i in range(1, int(n+1)):
        for j in range(0, int(i+1)):
            yield int(math.factorial(i) / (math.factorial(i-j) * math.factorial(j)))

pascal_triangle = new_num(numer)


def output(c,h):
    '''
    :param c: int, auxiliary variable;
    :param h: int, auxiliary variable;
    This function displays the coefficients in the form of a Pascal triangle
    '''
    for i in range(0, int((numer+1)*(numer+2)/2)):
        if i == int(c):
            print('')
            c = int(c + h)
            h += 1
        elif i != 0:
            print(" ", end="")
        print(next(pascal_triangle), end="")
output(1,2)
```

```
def check_num_page(num):
    '''
    :param num: int, number of pages;
    :return: number of pages
    '''
    try:
        num = int(input("Enter the number of pages: "))
        if 16 <= int(num) <= 1280 and int(num) % 16 == 0 or int(num) % 24 == 0 or int(num) % 32 == 0:
            return num
        else:
            raise ValueError
    except ValueError:
        print("Error, entering data incorrectly!")
        return check_num_page(num)
num_page = check_num_page('')

def check_num_notebook(num):
    '''
    :param num: int, number of pages in the notebook;
    :return: number of pages in the notebook.
    '''
    notebook = [16,24,32]
    lst = []
    for char in notebook:
        if int(num_page) % char == 0:
            lst.append(str(char))
    print(f"Select a number of pages: {' '.join(lst)}")
    try:
        num = int(input("Enter the number of pages in the notebook: "))
        if num in list(notebook):
            return num
        else:
            raise ValueError
    except ValueError:
        print("Error, entering data incorrectly!")
        return check_num_page(num)
num_note = check_num_notebook('')

def check_param(num):
    '''
    :param num: int, takes one or two;
    :return: one or two (boolean)
    '''
    try:
        num = int(input("Enter a 1 or 0: "))
        if 0 <= num <= 1:
            return num
        else:
            raise ValueError
    except ValueError:
        print("Error, entering data incorrectly!")
        return check_param(num)
num_param = check_param('')


def pro():
    '''
    This is the main function in which the pages are rearranged and the number of notebooks is determined.
    '''
    def profer_1(roster):
        points = []
        for lst in roster:
            n,m = 0, -1
            for i in range(2,10):
                if i % 2 == 0:
                    points.append(lst[m])
                    points.append(lst[n])

                else:
                    points.append(lst[n])
                    points.append(lst[m])
                m -= 1
                n += 1
                if n == -1 * m:
                    break
        return ', '.join(map(str,(points)))


    def profer_2(num):
        lst = []
        nero = []
        for char in range(1,num+1):
            lst.append(char)
            if len(lst) == 16:
                mainer = lst.copy()
                nero.append(mainer)
                lst.clear()
        return nero

    if num_param:
        print(profer_2(num_page))
        print(f"The number of notebooks in the book: {num_page // num_note} ")
    else:
        print(profer_1(profer_2(num_page)))
        print(f"The number of notebooks in the book: {num_page // num_note} ")

pro()
```
